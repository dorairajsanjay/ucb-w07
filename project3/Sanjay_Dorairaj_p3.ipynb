```python
X, Y = [], []

with open('./data/mushroom.data') as fdata:
    for line in fdata:
        items = line.split()
        Y.append(int(items.pop(0)))
        x = np.zeros(len(feature_names))
        for item in items:
            feature = int(item.split(':')[0])
            x[feature] = 1
        X.append(x)

# Convert these lists to numpy arrays.
X = np.array(X)
Y = np.array(Y)

# Split into train and test data.
train_data, train_labels = X[:7000], Y[:7000]
test_data, test_labels = X[7000:], Y[7000:]

original_X = X

# Check that the shapes look right.
print train_data.shape, test_data.shape
```

```python
# a handy function for vizualizing principle components as a heatmap
# this allows you to see what dimensions in the 'original space' are
# active

def visualize_pca_comps_heatmap(plot, comps):
    heatmap = plot.pcolor(comps, cmap=plt.cm.Blues)
    
    x_lab = [i for i in range(comps.shape[1])]
    y_lab = [i for i in (range(comps.shape[0]))]
    
    plot.set_xticks(np.arange(comps.shape[1])+0.5, minor=False)
    plot.set_yticks(np.arange(comps.shape[0])+0.5, minor=False)
    
    # want a more natural, table-like display
    plot.invert_yaxis()
    
    plot.set_xticklabels(x_lab, minor=False,)
    plot.set_yticklabels(y_lab, minor=False)
    
    plt.title('Heatmap of PCA components Rows: components, Cols: Original dimensions')
```

```python
def screeplot(pca, standardised_values,subplot_value=None,xlabel="x-axis",ylabel="y-axis"):
    y = np.std(pca.transform(standardised_values), axis=0)**2
    x = np.arange(len(y)) + 1
    if subplot_value is None:
        plt.plot(x, y, "o-")
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
    else:
        subplot_value.plot(x,y,"o-")
        subplot_value.set_xlabel(xlabel)
        subplot_value.set_ylabel(ylabel)
        

        
    #plt.xticks(x, ["C."+str(i) for i in x], rotation=60)
```

(3) Run KMeans with [1,16] clusters over the 2d projected data. Mark each centroid cluster and plot a circle that goes through the most distant point assigned to each cluster.

```python
#def P3():
### STUDENT START ###

import matplotlib
import seaborn as sns
from sklearn.datasets import make_swiss_roll
from sklearn.datasets import make_s_curve
from sklearn.datasets import make_circles

from sklearn.metrics.pairwise import euclidean_distances
from scipy.spatial.distance import cdist

class MyList(list):
    def __init__(self, *args):
        list.__init__(self, *args)
    def indices(self, filtr=lambda x: bool(x)):
        return [i for i,x in enumerate(self) if filtr(x)]

reduced_data = PCA(n_components=2).fit_transform(X)

# Split into train and test data after dimensionality reduction
train_data_pca = reduced_data[:7000]
test_data_pca  = reduced_data[7000:]

kmeans = KMeans(init='k-means++', n_clusters=16, n_init=10)
kmeans.fit(reduced_data)

matplotlib.rcParams.update({'font.size': 30})
plt.figure(figsize=(32,14))
plt.plot(reduced_data[:, 0], reduced_data[:, 1], 'k.', markersize=8,color='b')

#Plot the centroids as an X
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1],
            marker='X', s=300, linewidths=1,
            color='#141400', zorder=20,alpha=0.7)

plt.axes()

# initialize centroid
centroid = numpy.empty([1,2])

# find the farthest point from each centroid
for i in range(kmeans.n_clusters):
    
    my_list = MyList(kmeans.labels_)
    cluster_label_indexes = my_list.indices(lambda x: x==i)
    
    x_values = numpy.array(reduced_data[:,0])[cluster_label_indexes]
    y_values = numpy.array(reduced_data[:,1])[cluster_label_indexes]
    
    cluster = numpy.empty([len(x_values),2])
    
    cluster[:,0] = x_values[:,]
    cluster[:,1] = y_values[:,]
    
    centroid[0,0] = centroids[i,0]
    centroid[0,1] = centroids[i,1]
    
    dist = cdist(cluster,centroid,metric='euclidean')
    
    circle = plt.Circle((centroids[i,0],centroids[i,1]), radius=np.max(dist), color='black',fc='y')
    plt.gca().add_patch(circle)
    
plt.axis('scaled')

plt.title('K-means clustering on poisonous mushrooms dataset')
plt.show()

### STUDENT END ###

#P3()
```

(5) Fit two 4-component full covariance GMMs, one for the positive examples and one for the negative examples in your 2d projected data. Predict the test examples by choosing the label for which the model gives a larger probability (use GMM.score). What is the accuracy?

```python
#def P5():
### STUDENT START ###

import warnings
warnings.filterwarnings("ignore")

import matplotlib
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from sklearn import mixture

mpl.rcParams['agg.path.chunksize'] = 10000 
cov_mat_type = "full"

import matplotlib.pyplot as plt
from matplotlib import cm
from math import sin
import seaborn as sns

def gmm_contour_plot(clf,examples,plot_title):

    # Compute scores for the contour plot
    x = np.linspace(-4,4,75)
    y = np.linspace(-4,4,75)
    X, Y = np.meshgrid(x, y)
    XX = np.array([X.ravel(), Y.ravel()]).T
    Z = -clf.score_samples(XX)
    Z = Z.reshape(X.shape)

    # plot contours and scatter plot
    plt.figure(figsize=(30,15))
    CS = plt.contour(X, Y, Z, norm=LogNorm(vmin=1.0, vmax=1000.0),
                     colors=('r', 'g', 'b'),
                     levels=np.logspace(0, 3, 10))

    plt.scatter(examples[:, 0], examples[:, 1])

    title = plot_title,cov_mat_type
    plt.title(title,size=20)
    plt.axis('tight')

    plt.show()
    
def compute_components_accuracy(pca_components,gmm_components,cov_mat_type,gmm_plot=False,debug_enabled=False):
    
    reduced_data = PCA(n_components=pca_components).fit_transform(original_X)
    
    train_data_pca = reduced_data[:7000]
    test_data_pca  = reduced_data[7000:]

    positive_examples_indices = np.where(train_labels==1)
    negative_examples_indices = np.where(train_labels==0)

    positive_examples = numpy.array(train_data_pca)[positive_examples_indices]
    negative_examples = numpy.array(train_data_pca)[negative_examples_indices]

    # Fit a 4 component GMM for both and plot them    
    positive_clf = mixture.GaussianMixture(n_components=gmm_components, covariance_type=cov_mat_type)
    positive_clf.fit(positive_examples)

    negative_clf = mixture.GaussianMixture(n_components=gmm_components, covariance_type=cov_mat_type)
    negative_clf.fit(negative_examples)

    if gmm_plot is True:
        gmm_contour_plot(positive_clf,positive_examples,"Positive Examples - GMM Estimation. Components:4 Covariance Matrix Type:Full")
        gmm_contour_plot(negative_clf,negative_examples,"Negative Examples - GMM Estimation. Components:4 Covariance Matrix Type:Full")

    predicted_labels = np.zeros_like(test_labels)

    for i in range(1,len(test_data_pca)):

        pos_score = positive_clf.score(test_data_pca[i])
        neg_score = negative_clf.score(test_data_pca[i])

        if pos_score >= neg_score:
            predicted_labels[i] = 1
        else:
            predicted_labels[i]=0
            
    total_matches = (float)(np.count_nonzero(test_labels==predicted_labels))
    total_labels = (float)(len(test_labels))
    accuracy = total_matches/total_labels

    if debug_enabled is True:
        print "pca_components:",pca_components," gmm_components:",gmm_components," cov_mat_type:",cov_mat_type
        print "Total Matches:",total_matches
        print "Total Labels:", total_labels
        print "Accuracy:", accuracy
    else:
        print "pca_components:{0:2d} gmm_components:{1:2d} cov_mat_type:{2:10s} accuracy:{3:12.4f}".format(
            pca_components,gmm_components,cov_mat_type,accuracy)
    
    return accuracy

compute_components_accuracy(pca_components=2,gmm_components=4,cov_mat_type="full",gmm_plot=True,debug_enabled=True)

### STUDENT END ###

#P5()
```

(6) Ideally, we'd like a model that gives the best accuracy with the fewest parameters. Run a series of experiments to find the model that gives the best accuracy with no more than 50 parameters. For example, with 3 PCA components and 2-component diagonal covariance GMMs, you'd have:

( (3 mean vector + 3 covariance matrix) x 2 components ) x 2 classes = 24 parameters

You should vary the number of PCA components, the number of GMM components, and the covariance type.

```python
#def P6():

### STUDENT START ###

import numpy as np

# Search for optimal parameter mix
np.random.seed(0)

accuracy = 0
best_pca_components = 0
best_gmm_components = 0
best_cov_mat_type = ""

for pca_components in [5,10]:
    for gmm_components in [2,4,6,8]:
        for cov_mat_type in ['spherical', 'diag', 'tied', 'full']:
            new_accuracy = compute_components_accuracy(pca_components=pca_components,gmm_components=gmm_components,cov_mat_type=cov_mat_type)
            if (new_accuracy > accuracy):
                accuracy = new_accuracy
                best_pca_components = pca_components
                best_gmm_components = gmm_components
                best_cov_mat_type   = cov_mat_type
                
print "Best Values:"
print "pca_components:{0:2d} gmm_components:{1:2d} cov_mat_type:{2:10s} accuracy:{3:12.4f}".format(
    best_pca_components,best_gmm_components,best_cov_mat_type,accuracy)

                
### STUDENT END ###

#P6()
```
